```python
import os
from pymongo import MongoClient
from datetime import datetime
import uuid
```

```python
# Ruta principal del dataset
root = "trash_images"
```

```python
# Extensiones permitidas
valid_ext = (".jpg", ".jpeg", ".png")

for folder in os.listdir(root):
    folder_path = os.path.join(root, folder)

    if os.path.isdir(folder_path):
        print(f"Renombrando carpeta: {folder}")

        # Obtener archivos v√°lidos
        files = [f for f in os.listdir(folder_path) if f.lower().endswith(valid_ext)]
        files.sort()

        # Primera pasada: renombrar con nombres temporales para evitar conflictos
        temp_files = []
        for i, filename in enumerate(files, start=1):
            old_path = os.path.join(folder_path, filename)
            ext = os.path.splitext(filename)[1].lower()
            temp_name = f"temp_{i}{ext}"
            temp_path = os.path.join(folder_path, temp_name)
            os.rename(old_path, temp_path)
            temp_files.append(temp_name)

        # Segunda pasada: asignar nombre final
        for i, temp_name in enumerate(temp_files, start=1):
            ext = os.path.splitext(temp_name)[1]
            new_name = f"{folder}_{i:03d}{ext}"
            os.rename(
                os.path.join(folder_path, temp_name),
                os.path.join(folder_path, new_name)
            )

print("Renombrado con formato completo")
```
