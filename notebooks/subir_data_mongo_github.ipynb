```python
# ============================================================
# APARTADO A: SCRIPT COMPLETO PARA CREAR EL DATASET EN MONGODB
# A PARTIR DEL DATASET SUBIDO A GITHUB
# ============================================================

import os
import pymongo
from datetime import datetime
```

```python
# ======== CONFIGURACI√ìN ========
GITHUB_USER = "tigreraph"                 # tu usuario
REPO = "ecoflash-dataset"                # tu repositorio
BRANCH = "main"                           # rama principal

DATASET_PATH = "C:\\Users\\Jonna\\ecoflash-dataset"       # carpeta donde est√°n las im√°genes locales

MONGO_URI = uri = "mongodb+srv://jonnathanftigreest_db_user:j3nhScPaM7SpNacc@cluster0.lckuzqv.mongodb.net/?appName=Cluster0" # ruta mongo atlas
DB_NAME = "EcoFlash"
COLLECTION_NAME = "images"
```

```python
# ======== CONECTAR A MONGODB ========
client = pymongo.MongoClient(MONGO_URI)
db = client[DB_NAME]
collection = db[COLLECTION_NAME]

print("üîó Conectado a MongoDB Atlas correctamente.")
```

```python
docs_to_insert = []

for class_name in os.listdir(DATASET_PATH):

    class_folder = os.path.join(DATASET_PATH, class_name)

    # Verifica que sea una carpeta (cardboard, glass, etc.)
    if not os.path.isdir(class_folder):
        continue

    print(f"üìÇ Procesando categor√≠a: {class_name}")

    # Recorre las im√°genes dentro de la carpeta
    for filename in os.listdir(class_folder):

        file_path = os.path.join(class_folder, filename)

        # Ignora si no es archivo
        if not os.path.isfile(file_path):
            continue

        # ----------------------------------------------------
        # CREAR URL RAW DE GITHUB
        # ----------------------------------------------------
        raw_url = (
            f"https://raw.githubusercontent.com/"
            f"{GITHUB_USER}/{REPO}/{BRANCH}/{class_name}/{filename}"
        )

        # ----------------------------------------------------
        # DOCUMENTO PARA MONGODB
        # ----------------------------------------------------
        doc = {
            "filename": filename,
            "category": class_name,
            "filepath": f"{class_name}/{filename}",
            "url": raw_url,
            "uploaded_at": datetime.utcnow(),
            "source": "github_public_dataset"
        }

        docs_to_insert.append(doc)
```

```python
# ============================================================
# INSERTAR DOCUMENTOS EN MONGODB
# ============================================================

if docs_to_insert:
    # Antes de insertar, puedes limpiar la colecci√≥n si deseas evitar duplicados:
    # collection.delete_many({})  # ‚ö†Ô∏è OPCIONAL

    insert_result = collection.insert_many(docs_to_insert)

    print("\n============================================================")
    print(f"‚úÖ Se insertaron {len(insert_result.inserted_ids)} documentos en MongoDB.")
    print("üî• El dataset est√° listo para consultas y procesamiento.")
    print("============================================================\n")

else:
    print("‚ö†Ô∏è No se encontraron im√°genes en el dataset local.")
```
